```python
import pandas as pd
import numpy as np
from faker import Faker
from collections import defaultdict
from sqlalchemy import create_engine
import datetime
import random
import re
```

```python
from numpy import integer


fake=Faker()

Account= defaultdict(list)

row_number=int(input("Please write the number you wish to see the row size, DON'T CRASH ME!"))
```

```python
NAME_LIST =["EMRE","EFE","ERDEM","UYGAR","ÇAĞRI","MUHAMMED","SELCAN"]
SURNAME_LIST=["KAYA","KAPLAN","AVCI","BAYKARA","ER"]
BANK_LIST=["EU","İZM","İST","ANK"]
```

```python
for i in range (row_number):
    Account["SSN"].append(fake.ssn())
    Account["ADRESS"].append(fake.address())
   

    Account["FULL_NAME"].append(fake.name())
    Account["PHONE NUMBER"].append(fake.bothify(text='05#########'))
    
    Bank_Name="BANK-"+random.choice(BANK_LIST)
    if Bank_Name.split("-")[1]=="İZM":
        Account["ID"].append(fake.bothify(text='300####'))
    elif Bank_Name.split("-")[1]=="İST":
        Account["ID"].append(fake.bothify(text='100####'))
    elif Bank_Name.split("-")[1]=="ANK":
        Account["ID"].append(fake.bothify(text='200####'))
    else :
        Account["ID"].append(fake.bothify(text='900####'))
    
    Account["BANK-ACCOUNT"].append(Bank_Name)



```

```python
Money_Transfer= defaultdict(list)

Money_Transfer_Size=int(input("Please write the number you wish to see the row size, DON'T CRASH ME!"))
```

```python
for i in range (len(Account["ID"])): 
    max=random.randint(3,6)
    min=random.randint(1,3)
    random_size=random.randint(min,max)
    if len(Money_Transfer)==Money_Transfer_Size:
      break
    for j in range (random_size):

         
         random_ıd_choice=random.choice(Account["ID"])
         if Account["ID"][i]==random_ıd_choice:
            i=i-1
            break
         Money_Transfer["From_Bank_Id"].append(Account["ID"][i])
         Money_Transfer["To_Bank_Id"].append(random_ıd_choice)
         
         if random_size <= 3:
            Money_Transfer["AMAOUNT"].append(fake.pyfloat(right_digits=2,positive=True,min_value=20000.00,max_value=1000000.00))
         elif (random_size > 3) and (random_size <= 7)  :
            Money_Transfer["AMAOUNT"].append(fake.pyfloat(right_digits=2,positive=True,min_value=5000.00,max_value=20000.00))
         elif random_size  > 7  :
            Money_Transfer["AMAOUNT"].append(fake.pyfloat(right_digits=2,positive=True,min_value=1000.00,max_value=5000.00))
         Money_Transfer["TimeStamp"].append(fake.date_time_ad(start_datetime=datetime.datetime(2022,9,14)))
             
```
